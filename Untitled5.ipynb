```python
import numpy as np
from numpy.ma.core import shape

import matplotlib.pyplot as plt

from scipy.fft import fft, fftfreq
from scipy.io import wavfile
from scipy import signal as sig
from IPython.display import Audio
```

```python
def calc_spectr(signal, win):
  x = signal.copy()
  len_x = len(x)
  x = x - np.mean(x)
  if win is not None:
    x = x * win(len_x)
  X = fft(x, len_x)
  X = 2 / len_x * abs(X[0:len_x // 2])
  X[0] /= 2
  return X

def paint_spectr(signal, samplerate, xlim=None):
    spectr = calc_spectr(signal, None)
    f = np.linspace(0, samplerate // 2, len(spectr))
    s = 20 * np.log10(spectr)
    s[0] = 0.0
    plt.plot(f, s)
    if xlim is not None:
        plt.xlim(xlim)
```

```python
filepath = "./13.wav"
samplerate, x = wavfile.read(filepath)
x = x.mean(axis=1)

paint_spectr(x, samplerate, [0, 5000])
Audio(x, rate=samplerate)
```

```python
def ecvalayzer(x, samplerate, left, right, diff):
    ws = [left, right]
    wp = [int(left - samplerate * 0.01), int(right + samplerate * 0.01)]
    if wp[0] < 0.0:
        wp[0] = 0.0
    rp = 1
    rs = diff
    N, Wn = sig.cheb1ord(wp, ws, rp, rs, fs=samplerate)
    b, a = sig.cheby1(N, rp, Wn, 'bandstop', fs=samplerate, output='ba')
    filtered = sig.lfilter(b, a, x)
    return filtered
```

```python
# ,  ,   ,      
si = ecvalayzer(x, samplerate, 200, 600, 5)

paint_spectr(si, samplerate)
plt.xlim([0, 5000])
Audio(si, rate=samplerate)
```
